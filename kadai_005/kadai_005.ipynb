```python
import matplotlib.pyplot as plt
import numpy as np

# データの準備
# 1. 折れ線グラフ用
x1 = [0, 1, 2, 3]
y1 = [1, 2, -5, 2]

# 2. sin関数用
x2 = np.linspace(0, 10, 100)
y2 = 2 * np.sin(2 * np.pi * x2 / 3) + 2  # 振幅とオフセットを調整

# 3. ヒストグラム用（正規分布）
data3 = np.random.normal(15, 5, 2000)

# 4. 散布図用
x4 = np.random.normal(3, 2, 70)
y4 = np.random.normal(4, 2, 70)

# 描画キャンバスの作成 (2行3列)
fig, axes = plt.subplots(2, 3, figsize=(12, 8))

# --- 各グラフのプロット ---

# [0, 0] 折れ線グラフ
axes[0, 0].plot(x1, y1)

# [0, 1] sin関数
axes[0, 1].plot(x2, y2)
axes[0, 1].set_ylim(-1, 5)
axes[0, 1].set_xticks(range(10))

# [0, 2] ヒストグラム
axes[0, 2].hist(data3, bins=10)

# [1, 0] 散布図
axes[1, 0].scatter(x4, y4)

# [1, 1] と [1, 2] は空のまま（デフォルトで枠が表示されます）

# レイアウトの調整
plt.tight_layout()
plt.show()
```
