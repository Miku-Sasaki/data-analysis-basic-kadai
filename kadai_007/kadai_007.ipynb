```python
import pandas as pd
import matplotlib.pyplot as plt

# csvファイルをデータフレームとして読み込む
df = pd.read_csv("sample_pandas_6.csv")
# 先頭から5行目までを表示する
df.head()

df.query('商品番号 == "Z4WOOIYV"')
df.query('単価 == 600')
df.query('在庫 <= 5')
df.query('商品番号 == "8T7D5DQA" and 在庫 <= 5')

def tax(x):
    return x * 1.10

df['単価'].apply(tax)
type(df['発注日'].loc[0])

df['発注日'] = pd.to_datetime(df['発注日'])
type(pd.to_datetime(df['発注日']).loc[0])
df.agg({'発注日':['max', 'min']})

tax_series = df['単価'].apply(tax)
tax_series.name = "単価（税込み）"
pd.concat([df, tax_series], axis=1)

category_df = pd.read_csv('category.csv')
category_df

df = pd.merge(df, category_df[['商品番号', 'カテゴリー']], how='inner', on='商品番号')
df

# 1. カテゴリー列の出現頻度をカウントして棒グラフで表示
# 日本語フォントの設定（環境により文字化けする場合はここを調整してください）
plt.rcParams['font.family'] = 'MS Gothic'

df['カテゴリー'].value_counts().plot(kind='bar', title='カテゴリー別商品数')
plt.ylabel('件数')
plt.show()

# 2. 商品番号ごとの注文数の基本統計量を求める
# describe()を使うと、件数、平均、標準偏差、最小値、四分位点、最大値がまとめて出力されます
stats = df.groupby('商品番号')['注文数'].describe()

# 結果を表示
print(stats)
```
